```python
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

# defining image transformations

transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5], std=[0.5])
])

# loading datasets

train_data = datasets.ImageFolder('/path/to/dataset/train', transform=transform)
test_data = datasets.ImageFolder('/path/to/dataset/test', transform=transform)
val_data = datasets.ImageFolder('/path/to/dataset/val', transform=transform)

# creating dataloaders

train_loader = DataLoader(train_data, batch_size=32, shuffle=True)
test_loader = DataLoader(test_data, batch_size=32, shuffle=False)
val_loader = DataLoader(val_data, batch_size=32, shuffle=False)
```
